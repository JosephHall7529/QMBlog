# Solution of Klein-Gordon Equation for a square well potential

## Problem
Find the solution of the KG equation for the π meson in a Coulomb potential and discuss the energy eigenvalues. The pion has ass 139.577 MeV and Spin 0. It therefore obeys the KG equation.

We can rearrange the KG equation with the inputted square well potential and arrive at a Bessel like differential equation:

- We omit the neumann sfunction solution diue to its singular nature at the origin 

When we consider the complex equivalent of this differential equation we find the solution takes a Hankel form.
